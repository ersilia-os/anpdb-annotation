```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.hist(df["ra_score_eos2r5a"])
plt.title("Retrosynthetic Accessibility Score")
```

```python
plt.hist(df["sa_score_eos9ei3"])
plt.title("Synthetic Accessibility Score")
```

```python
plt.hist(df["np_score_eos9yui"])
plt.title("Natural Product Score")
```
